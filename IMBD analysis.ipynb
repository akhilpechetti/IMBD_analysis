## Exploratory Analysis and Visualization

Let's begin by importingmatplotlib.pyplot and seaborn.


```python
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('dark')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
sns.countplot(df['Year'])
plt.show()
```

By this we can observe that the number of movies increased year by year and there was a rapid growth in 2016

```python
plt.figure(figsize=[14,8])
plt.hist(df['Rating'])
plt.xticks(np.arange(0,11))
#print(df['Rating'].value_counts())
#plt.show()
min_rating=print("MIN RATING =",df['Rating'].min())
max_rating=print("MAX RATING =",df['Rating'].max())
df['Rating'].mean()
```

```python
df['Rating type']="unknown"
df.loc[df['Rating']<6,'Rating type']='Good'
df.loc[(df['Rating']>=6)&(df['Rating']<8),"Rating type"]="Better"
df.loc[(df['Rating']>=8),"Rating type"]="Best"
df.head(6)
```

```python
sns.countplot(df['Rating type'])
plt.title("Movie type")

plt.show()
```

```python
plt.figure(figsize=[20,5])
sns.barplot(x="Rating",y='Metascore',data=df)
plt.show()
```

```python
#df['Runtime (Minutes)'].min()
#df['Runtime (Minutes)'].max()
sns.distplot(df['Runtime (Minutes)'])
plt.title('Distribution of Movie Time')
plt.ylabel('Movie time(min)')
print('Mean movie Time=',df['Runtime (Minutes)'].mean(),'min')
plt.show()
```

we can observe the mean time for a movie comes to be 144 min

```python
sns.violinplot(x='Rating type',y='Rating',data=df)
plt.show()
```

```python
ak=list(df['Genre'])
ak[0:5]
```

```python
Horror=df.loc[df['Genre'].str.contains('Horror')]['Rank'].count()
Action=df.loc[df['Genre'].str.contains('Action')]['Rank'].count()
Adventure=df.loc[df['Genre'].str.contains('Adventure')]['Rank'].count()
Mystery=df.loc[df['Genre'].str.contains('Mystery')]['Rank'].count()
Scific=df.loc[df['Genre'].str.contains('Sci-Fi')]['Rank'].count()
Fantasy=df.loc[df['Genre'].str.contains('Fantasy')]['Rank'].count()
Family=df.loc[df['Genre'].str.contains('Family')]['Rank'].count()
Comedy=df.loc[df['Genre'].str.contains('Comedy')]['Rank'].count()
Crime=df.loc[df['Genre'].str.contains('Crime')]['Rank'].count()
Biography=df.loc[df['Genre'].str.contains('Biography')]['Rank'].count()
Drama=df.loc[df['Genre'].str.contains('Drama')]['Rank'].count()
Thriller=df.loc[df['Genre'].str.contains('Thriller')]['Rank'].count()
Genre_df=pd.DataFrame({'Genre':['Horror','Action','Adventure','Mystery','Scific','Fantasy','Family','Comedy','Crime','Biography','Drama','Thriller'],'No:of movies':[Horror,Action,Adventure,Mystery,Scific,Fantasy,Family,Comedy,Crime,Biography,Drama,Thriller]})
Genre_df
```

```python
plt.figure(figsize=[12,8])
sns.barplot(Genre_df['Genre'],Genre_df['No:of movies'])
plt.show()
```

From the above visuvalization we can observe that the viewers are more attracted towards Drama films. so,there were more number of movies targeting Drama

```python
fig = plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(x="Rating type",hue="Year",data=df)
plt.title("Rating type results every year")
plt.subplot(1, 2, 2)
sns.countplot(x=(df.Votes>140000),hue=df.Year)
plt.show()
    
```

```python

sns.set_style('whitegrid')
sns.countplot(x=(df.Rating>6),hue=df.Year)
plt.title('Rating through years')
#sns.catplot(x=df.Rating>6, hue="Year", col="Rating type",data=df, kind="count",height=4, aspect=.7)
plt.show()
```

We can observe that ratings were increased through the years

By this we can say that people are showing more intrested towards movies in present compared to past
